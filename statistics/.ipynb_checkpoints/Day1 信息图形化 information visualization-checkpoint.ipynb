# 信息图形化（ information visualization）

## 第一印象（First image）

- 在位手头数据无法给出事情的真相而发愁么？

统计能够化繁为简，然你从一堆令人困惑的数据中找到真相。借助可视化的方法，使之公之于众。

**统计**,是通过某种有意义的方式对原始事实和数字进行提炼，使得仅仅通过观察原始数据无法立即水落石出的一些理念得以昭示。

统计的研究包括：

- 统计数据的来源
- 计算方法
- 使用有效方法得出结论

**统计既能得到真相，也能给出谎言**

- 为什么要使用图形化？

只看原始数据无法明白，数据中隐含的模式和趋势。图形是发现隐藏模式的有效方法（对人来说）。

- 信息和数据有何差别？

数据值得是所搜集的原始事实与数字。信息指的是加入了某种意义的数据。

- 制图软件能解决所有问题么？

制图软件可以节省时间，但无法替你思考。

## 各种图形

### 饼图

饼图为一个园，将数据分为几类，每一块扇形表示一个类或组。扇形的大小表示这类数据栈总体的比例。

扇形越大，该类的相对频繁程度越大。一个特定组中的对象数目称为**频数**。

- 什么时候有用？

每个扇形块的大小，表示每组数据的相对频数。所以在对基本比例进行比较的时候，饼图有用。很容易从中发现哪个组的频数较高。

饼图的作用是对不同组/类所占的比例进行比较。

- 什么时候没用？

当各组之间的相对频数差别不大时，用处不大。

## 条形图

条形图能够比较相对大小，而且还具有精确性。

- 什么时候用？

当各类数据频数大致相同时，条形图更为合理。

能够发现细小的差别。

### 垂直条形图

横轴表示类，纵轴表示频数或百分数。每个长方形表示一类，长方形的高度代表相应类的数值。

更为常见

### 水平条形图

纵轴表示类，横轴表示频数或百分数。每个长方形表示一类，长方形的高度代表相应类的数值。

类名称太长时，适用。

### 标度

可以从0开始，也可以不从0开始计数。**参照点不同会让人有不同的感觉。**

### 堆积条形图

为了能够在一张图上显示多批数据，例如满意的和不满意的人数，可以使用多个条形，每个条形显示一种数据。

- 什么时候有用？

相比较多个频数时。

### 分段条形图

每个条形表示一个类别，但按比例将这个长方形分割为几段，整体长度表示整体频数，每段表示一个细分类型的频数。

例如：体育游戏用户是一各类别，满意的人数和不满意的人数是两个子类。这时，既能看到频数，也能看到百分比。

- 什么时候有用？

同时体现频数和百分数，可以使用。



# 统计量

频数：一个特定组中的对象数目称为**频数**。

- 图形化时，频数标度什么时候用频数？什么时候用百分数？

这取决于你想传达什么信息。

记住一条黄金定律：**不管是否选择百分数，都要在统计中列出频数，或是将频数标在图形中间/旁边**。

这样可以让人们知道样本量。

**要特别当心：只有百分数没有频数，或只有频数而没有百分数，都是对读者的愚弄**。
