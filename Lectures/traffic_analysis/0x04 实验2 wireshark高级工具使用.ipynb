# wireshark高级工具使用

上面的实验主要为了向大家介绍wireshark的基本使用，使大家对wireshark有基本的了解。下面将对一些功能做进一步的详细介绍。

## 标记与打印数据包

在今后学习或工作过程中，我们会遇到大量网络数据包，当其数量很大时，需要使用高效的分析方法。

wireshark提供了一些功能，能够将符合条件的数据包（关键信息）查找出来，加注标记。

### 查找数据包

在前面的初步使用中已经介绍过，我们不仅可以使用菜单“编辑”-“查找分组”打开查找功能，也可以使用“CTRL+F”快捷键打开查找对话框。

如果想找下一个满足条件的数据包记录，可以按快捷键“Ctrl+N”；想找前一个，可以按快捷键“Ctrl+B”。

查找功能与显示过滤器的不同在于，查找时不会隐藏不符合条件的数据包，而显示过滤器会隐藏那些不符合条件的数据包。

在执行查找时，可以指定查找的窗口范围，有3种选择：

- 分组列表
- 分组详情
- 分组字节流

此外，还可以指定字符集，有3中选择：

- 宽窄
- 窄（UTF-8/ASC-II)
- 宽（UTF-16)

![标记数据包](images\03\查找细节选项.png)


### 标记数据包

在找到符合条件的数据包后，可以对这些记录进行标记。

右键点击要标记的数据包记，选择“标记/取消标记 分组”，或者在左键点击要标记的数据包记录后，按快捷键“Ctrl+M”。标记后的记录将以黑底白字高亮显示。如下图所示。

![标记数据包](images\03\标记数据包.png)

取消标记的操作如标记操作相同，这是一个类似开关的操作。

如果需要在已经标记的记录间快速切换，可以使用快捷键“SHIFT+CTRL+N”，移到下一个记录；使用快捷键“SHIFT+CTRL+B”，移到前一个记录。

值得一体的是，当希望需要选择多条记录时，wireshark却没有多选操作方式。我们需要通过左键点击每一条兴趣记录，然后进行标记，这样重复多次，完成对一组兴趣数据包记录的标记。

### 打印数据包

有的时候我们需要将数据包信息打印出来，可以通过菜单“文件”-“打印”打开打印对话框。

![打印对话框](images\03\打印对话框.png)

我们可以将选中的分组或全部分组按照显示详情或展开全部详情的方式打印出来。

如果没有合适的打印机，可以先将数据打印为pdf文件。

## 设定时间显示格式和相对参考

时间在数据包的分析中十分重要。

所有在网络上发送的事情都是与时间有关的。例如，我们通过分析数据包间存在的时间规律和网络延迟。

我们下面介绍时间的显示格式和相对参考。

### 时间显示格式

Wireshark所捕获的每个数据包都会由操作系统给与一个时间戳。

Wireshark可以显示这个数据包被捕获时的绝对时间戳，也可以显示与上一个被捕获的、捕获开始、捕获结束等数据包相对的时间戳。

在菜单“视图”-“时间显示格式”中有多个与时间显示相关的选项。

![(时间显示选项](images\03\时间显示选项.png)

在这个选项表中，可以设置：

- 显示的格式
- 显示的精度

### 数据包的相对时间参考

设置数据包的相对时间参考，可以让用户以一个数据包作为基准，而之后的数据包都以此计算相对时间戳。当一系列的顺序事件不是在捕获开始时被触发，而是在中间某地方被触发，这个功能将非常有用。

例如，如果在局域网内存在病毒感染，我们可以在交换机上捕获了局域网内的特殊网络数据，通过分析数据包间的时间关系，发现恶意代码的传播规律。

通常，在进行分析前，我们会选定一个数据包作为时间参考。

在数据包列表（Packet list）面板中选择一个作为基准的数据包，然后在菜单“编辑”-“设置时间参考（Set time reference）”或者按快捷键Ctrl+T，设置后这条数据包记录的时间戳将改变为“REF”。

![(时间显示选项](images\03\设置时间参考.png)

注意，这时我们要把时间显示格式设置为“自捕获开始经过的秒数”才有利于我们理解。

- 首先下载并打开样本文件：lotsofweb.pcapng。
- 在打开的数据包记录文件中，尝试使用“字符串”查找方式，在“分组详情”面板中查找包含“www.cnn.com” 记录

## 使用过滤器

Wireshark有两种过滤器：

- 捕获过滤器：当进行数据包捕获时，只有那些满足给定条件的才会被捕获。
- 显示过滤器：用于对已经捕获的数据包集合中，过滤出用户希望当前显示的数据包，隐藏当前不关注的数据包。

### 捕获过滤器中的BPF语法

捕获过滤器的基本情况之前已经介绍过了。这里重点介绍没介绍过的过滤器BPF语法。

在捕获过滤器的过滤条件使用的就是BPF语法，如图所示：

![捕获过滤器BPF语法](images\03\捕获过滤器BPF语法.png)

BPF（Berkeley Packet Filter）语法是用于Winpcap的过滤数据包的方法。

很多数据包嗅探软件都依赖于使用BPF进行捕获过滤。例如使用了libpcap库或winpcap库等等。

使用BPF语法创建的过滤器被称为表达式，并且每个表达式包含一个或多个原语。每个原语包含一个或多个限定词，然后跟着一个ID名字或数字，如下表所示：

|限定词|说明|例子|
|-|-|-|
|Type|指出名字或数字代表的意义|host、net、port|
|Dir|指明传输方向是目的还是来源|src、dst|
|Proto|限定所要匹配的协议|ether、ip、tcp、udp、http、ftp|

例如：设定捕捉来自（src）主机（host）192.168.0.10的tcp 端口80的数据包
```
src host 192.168.0.10  && tcp port 80

```

### BPF过滤条件

主要有以下几种过滤条件：

- 主机名和地址

使用过滤器时通常都会关注一个或一些特定的网络设备。根据网络设备的MAC地址、IP地址、DNS主机名来设置过滤条件是最常使用的过滤条件。

例如：
```
#使用host限定词定向捕获某个ip地址往来的流量
host 172.16.16.149
#使用host过滤某个ipv6地址
host 2001：db8:85a3:8a2e:370:7334
#使用主机名过滤
host testserver
#使用src限定流向
src host 172.16.16.149
#使用协议限定词ether
ether host 00-1a-a0-52-e2-a0
```

- 端口

在分析某些应用程序时，往往会使用端口号作为过滤条件

例如:
```
#只捕获8088端口的流量
port 8088

#捕获除8088端口以外的流量
!port 8088

#捕获目的地为端口8088的流量
dst port 8088

```

- 协议

当分析IP地址、端口号不能简单得到结论时，往往要综合协议类型进行分析。

以协议类型为过滤条件是较为常见的过滤设置。

```
#仅捕获ICMP流量
icmp
#捕获除ICMP流量以外的所有协议流量
!icmp
```
- 协议域

BPF过滤器和wireshark过滤器都提供了细粒度过滤，即允许用户根据协议头中的任何一个字节的值来创建过滤条件。协议域过滤条件指的就是可以针对某个协议封装中的具体字节值作为条件，进行捕获或显示过滤。

例如，我们想基于ICMP过滤器的类型域进行过滤，而类型域位于数据包的最开头，也就是偏移量为0的位置，可以在协议限定符后输入由方括号括起来的偏移量获取相应二级制数据（即icmp[0]），然后根据该值进行过滤。
![ICMP报文](images\03\ICMP报文.png)
![ICMP报文类型](images\03\ICMP报文类型.png)


捕获条件设置如下：

```
#仅捕获类型为3的ICMP包
icmp[0] == 3
```
### 显示过滤器

显示过滤器应用于已捕获文件中的显示设置。通过使wireshark应用显示过滤条件，在数据包列表中隐藏那些不符合条件的记录，仅显示符合条件的数据包记录，可以使用户的注意力更加聚焦。

我们通常会用显示过滤器清楚掉不相关的广播流量，例如隐藏与当前问题无关的ARP广播和DNS查询：

显示过滤器举例noarpandnodns](images\03\显示过滤器举例noarpandnodns.png)

### 显示过滤器的过滤表达式

在显示过滤器输入框的右侧，有一个“表达式...”按钮，点击后会弹出过滤表达式对话框。

![显示过滤器过滤表达式对话框](images\03\显示过滤器过滤表达式对话框.png)

过滤表达式对话框主要用于帮助用户建立显示过滤器（显示过滤表达式），它是一个关系或逻辑表达式。

下面介绍一下对话框内的组成：

- 左侧的“字段名称”框中显示了可以选择的过滤条件，主要是一些协议名称和可选用的细节选项。有面向对象编程经验的同学，可以认为每个协议是一个类，而细节选项就是这个类的属性。

- 左侧下方的“搜索”可以用来快速定位“字段名称”，例如输入IPv4可以帮助我们找到常见的IP协议。

- 右侧的“关系”框中，列举了可用的关系运算符。

- 右侧的“值”输入框，可以输入具体的条件值。

- 右侧的“范围（偏移：长度）”用于指定偏移位置。

- 下方的输入框中是书写表达式的地方

这个工具的具体使用步骤如下：

- 在“搜索”框中键入希望作为过滤条件一部分的协议名，或部分名称，这样可以缩小“字段名称”中的可选范围，有助于快速找到具体协议。例如键入ICMP；

- 点击“字段名称”中的某个需要设置过滤条件的协议名，例如ICMP，这时会弹出一个子层次，这些子层次中的内容可以帮助用户将协议的局部组成作为过滤条件。例如：选择icmp.type，此时，下方的输入框中会有相应的关键字出现；

- 选择“关系”中的某个关系，例如“==”；

- 在“值”中键入“3”；

- 点击“OK”，完成显示过滤表达式的设置。

## 名字解析

网络数据在发送接收时，使用了各种字母数字组成的地址，例如：链路层的MAC地址、网络层的IP地址、传输层的端口地址、某些应用层协议中的id号等等。

这些地址通常比较长且难记，名字解析，或称名字查询就是一个用来将这些地址转换为人们熟悉的字符串的协议或标准程序。

事实上，我们可以到下列地址人工查询各类型的地址：

- 对于MAC地址，可以到IEEE网站查询注册信息。

凡是希望自己的产品能够在国际市场上流通的组织，都会到IEEE进行注册。但有些单位在IEEE注册后，缴纳了保密费用，是查不到的。

IEEE 注册信息查询地址：https://regauth.standards.ieee.org/standards-ra-web/pub/view.html#registries

- 对于IP地址，可以到IANA或IANA的区域注册RIR进行查询，或者直接使用互联网上一些机构提供的服务，例如：https://geoip.com/ 。

- 端口地址，可以到IANA或IANA的区域注册RIR进行查询，或者使用 http://tool.oschina.net/commons?type=7 等互联网服务。

- 域名地址，可以到IANA或IANA的区域注册RIR进行查询，或者使用 http://whois.chinaz.com/ 等工具。

### 名字解析的问题

名字解析有其优点，使人们更容易理解数据，但也存在一些问题：

- 名字解析可能会失败，尤其是当我们查询的名字服务器不了解当前数据时；

- 名字解析在每次打开捕获文件时都要重新计算，而如果此时无法连接解析服务器就会解析失败；

- 对DNS解析时，会产生额外的数据包，即在捕获文件中会加杂着wireshark产生的流量，这事实上是一种干扰；

- 名字解析会增加系统资源的消耗，尤其是对记录大量流量的任务。

## 协议解析

wireshark中的协议解析器可以使数据包拆分成多个协议区段，以便分析。

wireshark为了能够解析协议，它拥有内置的解析器，但我们可以使用自己指定的解析器代替它。

### 更换解析器

wireshark的协议解析工作一般都是正确的，但不是绝对的。尤其是网络上的协议使用了非标准配置，例如端口变更。

请大家打开样例库中的wrongdissector.pcap样本文件。

这个文件中包含了大量SSL通信。SSL是安全套接层协议（Secure Socket Layer protocol），用来在主机间进行安全加密传输。

由于其保密性，所以大多数情况下，wireshark查看到的ssl流量是乱码。但在我们打开的样本中，我们可以看到一些明文信息。

事实上，在第4个数据包中，我们通过packet bytes面板能够看到filezilla（这是一种FTP服务产品），后面的数据包甚至包含了用户名和密码请求。

真的SSL流量是不可能看到明文的，所以这一段是被wireshark误以为是SSL的FTP流量。原因是，这个FTP流量使用了原本用于HTTPS的443端口。

为了解决这个问题，可以强制wireshark对这个数据包使用FTP协议解析器。这被称为**强制解码**。

#### 本节实验具体操作

- 右键单击一个SSL数据包，选择“Decode as...”或“解码为”；
- 选择"当前"为FTP。

## 数据包长度

数据包的长度（大小）可以给我们以丰富的信息。

正常情况下，以太网帧的最大长度为1518字节（Bytes），去掉帧中以太网（头14Bytes、尾4Bytes）、IP（20 Bytes）、TCP协议的头部（基本部分20 Bytes），还剩下1460字节。

帧的最大长度 $ 1518字节 = 14 + 20 + 20 + 1460 + 4 $

这1460字节是可以供给应用层协议或数据使用的最大字节数。

我们可以使用wireshark提供的分组长度统计功能对当前捕获的数据包进行概要分析。

例如，我们先打开样本文件download-slow.pcap，然后选择菜单“统计”-“分组长度”。如图所示：

![分组长度统计](images\03\分组长度统计.png)

不难发现：

- 大小为1280-2559字节的数据包占据了66.4%的带宽，这种大数据包通常是传输数据使用的。

- 大小为40-79字节的数据包占据了33.4%的带宽，这种小数据包，通常是用于传输控制信息的，不包含数据。
  + 以太网头尾共18字节
  + IP头最小20字节
  + 没选项的TCP头20字节
  + 常见TCP控制包：SYN、ACK、RST、FIN等
 
- 这段捕获记录，显然在进行数据下载或上传。

### 本节实验具体操作

- 在样本库中找到download-slow.pcap文件，打开它；
- 选择菜单“统计”-“分组长度”，查看分组长度统计窗口；

## 专家信息

专家信息用于揭示状态和诊断故障。

有4类专家信息：

- 对话（chat）：关于通信的基本信息；
- 注意（Note）：正常通信中的异常数据包；
- 警告（Warning）：不是正常通信的异常数据包；
- 错误（Error）：数据包中的错误，或解析器解析错误。

启用这一功能，需要打开菜单“分析”-“专家信息”。
