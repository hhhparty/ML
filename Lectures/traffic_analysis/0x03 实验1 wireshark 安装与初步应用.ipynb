## Wireshark常用图表

Wireshark提供了多种统计分析工具，这些统计分析结果将对日后的网络流量分析有非常大的帮助。

### 使用wireshark查看端点（Endpoint）信息

我们首先来了解端点分析工具，启动端点统计的方法是依次点击菜单中的“统计”-“端点”。

有必要解释一下“什么是端点？”，

在本门课程中，端点指具有网络地址的设备或软件，它们可能是某台主机、交换机、路由器、或运行于物理设备上的软件，有的以MAC地址标识，有的以IP地址标识，有的以端口地址标识，我们把这些网络数据包的发送点或接收点统称为“端点”。

下面图片显示了之前捕获结果的端点统计。

wireshark的端点统计是结合OSI网络体系模型，按网络协议分类显示的。例如：下图中按“以太网、IPv4、IPv6、TCP、UDP”等多个协议统计了当前打开或捕获的数据包集合中的端点统计结果。

![wireshark的端点统计表格](images\03\capture-wlaq01-http-endpoint.png)

一般来说，wireshark会自动列出当前数据集中所有的与端点统计有关的协议，但如果用户需要查看更多的协议，可以点击右下角的“Endpoint 类型”按钮，选择协议类型。

统计表中的每一行，表示对以某个网络地址（MAC地址、IP地址、端口地址）标识的端点进行统计：

- Address：表示地址值
- Packets：表示该地址收发的包的总数
- Bytes：表示该地址收发的总字节数
- Tx Packets：表示该端点向外发送的包的总数
- Tx Bytes：表示该端点向外发送的字节的总数
- Rx Packets：表示该端点接收到的包的总数
- Rx Bytes：表示该端点接收到的字节的总数

端点统计表的左下角有“解析名称”的勾选框，选中后wireshark将会对MAC地址进行解析，即查阅已在IEEE注册且公开信息的MAC地址块的所属单位的信息。

用户可以对端点统计结果进行复制，导出为csv文本表格文件或YAML文件，以便后续分析使用。

### 使用wireshark查看会话（Conversation）信息

启动“会话”或“对话”统计的方法是依次点击菜单中的“统计”-“会话”或“对话”。

会话统计同端点统计类似，也是结合OSI网络体系模型，按网络协议分类显示的。

所谓“会话”或“对话”，就是网络通信的收、发两端之间的通信过程，我们把发出数据包的一端称为“信源”或“源”，接收数据包的一端称为“信宿”或“目的”。在下面的会话统计窗口列表中，以“Address A”和“Address B”表示。

![wireshark的端点统计表格](images\03\capture-wlaq01-http-conversation.png)

Wireshark的会话统计对“A到B”、“B到A”的数据包的数量、字节数、持续时间等进行统计，这有利于我们发现网络通信中哪些会话是当前网络中的主要流量，哪些是异常流量，哪些属于异常协议通信等等。

仔细观察上图，会发现在IPv4面板中，有关城市、国家的信息是空白的，这些信息能够获得呢？

当然是可以的，打开菜单“编辑”-“首选项”-“Name resolution”,编辑“maxmid database directory”，会允许你键入一个目录地址，这个地址下需要下载IP地址与物理地址的对应关系数据库。

老版本的wireshark使用.dat的数据库，新版本的要求使用.mmdb数据库。免费的GeoIP数据库（.mmdb数据库）可以从https://dev.maxmind.com/geoip/geoip2/geolite2/ 下载。

### wireshark的协议分层统计窗口

wireshark的协议分层统计窗口显示了当前数据文件中所有的协议，及其树状分用情况。

![wireshark的协议分级统计](images\03\协议分级统计.png)

这个窗口是我们进行网络数据分析时最常用到的统计窗口，通过这个窗口能够清楚的了解当前网络中的活跃协议类型，各种通信协议的占比，给我们最为便捷和直观的网络流量概貌。

使用协议分层统计窗口，结合实际工作经验，可以使我们快速判断使用网络的部门，例如：
- IT管理部门的网络中，通常会有ICMP和SNMP流量
- 订单管理部门的网络中，通常会有SMTP流量
- ARP流量一般不会超过网络流量的10%
