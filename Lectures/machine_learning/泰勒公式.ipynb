# 函数的微分

## 微分的定义与计算

设函数$y=f(x)$定义在某区间有定义，$x_0$及$x_0 +\Delta x$在这区间内，如果函数的增量$\Delta y = f(x_0 + \Delta x) - f(x_0)$ 可表示为 $\Delta y = A \Delta x + o(\Delta x)$

其中A是不依赖于$\Delta x$的常数，那么称函数$y=f(x)$在点$x_0$是可微的，而$ A \Delta x$叫函数$y=f(x)$在点$x_0$相应于自变量增量$\Delta x$的微分，记作$dy$,即$dy = A \Delta x$。

那么，我们自然要问“在什么条件下，函数$y=f(x)$在点$x_0$可微？”

函数$y=f(x)$在点$x_0$可微 $\leftrightarrow$ 函数$y=f(x)$在点$x_0$可导，且当函数$y=f(x)$在点$x_0$可微时，其微分一定是$dy = f^{'}(x_0) \Delta x$。

（1）函数$y=f(x)$在任意点x的微分称为函数的微分，记作$dy或df(x)$，即$dy = f^{'}(x)\Delta x$；

（2）把自变量的增量$\Delta x$称为自变量的微分，记作$dx$ ,即$dx = \Delta x$

$dy = f^{'}(x)dx=\frac{dy}{dx}$，所以导数又称为微商。

**我们可以利用导数，研究函数的性质，解决实际问题。这就需要先了解导数与函数的关系，反映这个关系的有3个定理。**

- 罗尔定理
- 拉格朗日中值定理
- 柯西中值定理

## 费马引理

设函数$f(x)$在点$x_0$的某邻域$U(x_0)$内有定义，并且在$x_0$处可导，如果对任意$x \in U(x_0)$,有 $f(x) \leq f(x_0)$或 $f(x) \geq f(x_0)$，那么$f^{'}(x_0)= 0$.

注意：

- 导数为0的点为函数的驻点，也称为稳定点、临界点；
- 若函数在$x_0$处可导，且$f(x_0)$是$f(x)$在某邻域$U(x_0)$上的最大/小值，那么$f^{'}(x_0)= 0$
- 这个定理的逆定理不一定成立。例如，$y=x^3$在x=0处，就不是极值，但导数为0.

## 罗尔定理

如果函数$y = f(x)$满足：

- 在闭区间$[a,b]$上连续；
- 在开区间$(a,b)$内可导；
- 在区间端点处的函数值相等，即$f(a) = f(b)$；

那么在$(a,b)$内至少有一点$\xi (a < \xi < b)$，使得$f^{'}(\xi)= 0$

## 拉格朗日中值定理

如果函数$y = f(x)$满足：

- 在闭区间$[a,b]$上连续；
- 在开区间$(a,b)$内可导；

那么，在$(a,b)$内至少有一点$\xi (a < \xi < b)$，使得$f(b)-f(a) = f^{'}(\xi)(b-a)$

注意：

- 若$f(a) = f(b)$,则拉个朗日中值定理就是罗尔定理。
- $\frac{f(b)-f(a)}{(b-a)} = f^{'}(\xi)$，几何意义是在连续曲线f(x)（曲线弧AB），其除端点处处处有不垂直于x轴的切线，那么在此曲线弧上至少有一点c，在c处的切线于弦AB平行。
- 只有满足两个条件，定理才成立。

## 柯西中值定理

设$f(x),g(x)在[a,b]$上连续，在$(a,b)$上可微，且$g^{'}(x) \neq 0$对$(a,b)$中所有x成立，则一定存在$x \in (a,b)$使$\frac{f(b)-f(a)}{g(b)-g(a)} = \frac{f^{'}(\xi)}{g^{'}(\xi)}$

## 泰勒展开式

数学家泰勒想找到一种函数替代方法，使一个函数近似表达为另一个形式。

例如，近似表达$cosx,sinx e^x$.

泰勒认为，仿照一段曲线（使用g(x)仿照f(x)），要先保证起点相同，再保证此处导数相同，再保证此处导数的导数相同...

这要求，新函数g(x)必须能够无限求导；但是现实中不能无限求导，那么应该至少求多少次？

泰勒不想使用三角函数、指数函数的，因为它们是不好求值，所以，泰勒使用了多项式$g(x) = a_n x^n + a_{n-1} x^{n-1} + ... + a_{1} x^{1} + a_{0}$

然后再下列式子成立时求出系数$a^n$:

- $f(x_0) = g(x_0)$
- $f^{'}(x_0) = g^{'}(x_0)$
- $f^{''}(x_0) = g^{''}(x_0)$
- ...
- $f^{(n)}(x_0) = g^{(n)}(x_0)$

因为

$f^{(n)}(x_0) = g^{(n)}(x_0) \rightarrow a_n = \frac{f^{(n)}}{n!} $

所以，可以得出下列等式：

**若函数$f(x)$在$x_0$的某个闭区间$[a,b]$内都有n阶导数，且在开区间$(a,b)$上具有(n+1)阶导数，则对于闭区间$[a,b]$上任意一点x，有下列式子成立：**

$f(x） \simeq g(x) = g(x_0) + \frac{f^{(1)}(x_0)}{1!}(x-x_0)^{1}+...+\frac{f^{(n)}(x_0)}{n!}(x-x_0)^{n}$

在实际操作时，可以根据精度要求选则n值，只要n不是正无穷，那么一定要保留上式的约等于号。

如果想要把近似等于号拿掉，可以增加余项：

**佩亚诺（Peano）余项**

$R_n(x) = o[(x-x_0)^n]$

这里仅需要n阶导数存在。

**拉格朗日（Lagrange）余项**

$R_n(x) = f^{(n+1)}[x_0 + \theta (x-x_0)] \frac{(x - x_0)^{n+1}}{(n+1)!}$

这里仅需要n阶导数存在。
