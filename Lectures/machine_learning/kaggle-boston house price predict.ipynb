## 数据清洗 (Data Cleaning)

这里主要的工作是处理缺失值，首先来看各特征的缺失值数量：

查看哪些属性中存在空值，并且确定空值计数。

```python
aa = train.isnull().sum()
#print(aa)
aa[aa>0].sort_values(ascending=False)
```

考虑到有些房子没有地下室、没有单板区，所以用“None”和“0”填充这些字符串或数值区。

```python
strColumns = ["BsmtFinType2" , "BsmtExposure", "BsmtFinType1", "BsmtCond", "BsmtQual", "Electrical"]

for col in strColumns:
    train[col].fillna("None", inplace=True)
    
numColumns = ["MasVnrArea"]
for col in numColumns:
    train[col].fillna(0, inplace=True)
```

### 离散型变量的排序赋值

对于离散型特征，一般采用pandas中的get_dummies进行数值化，但在这个比赛中光这样可能还不够，所以下面我采用的方法是按特征进行分组，计算该特征每个取值下SalePrice的平均数和中位数，再以此为基准排序赋值，下面举个例子：

Foundation 这个特征表示房子的类型，将数据按其分组：

```python
train.groupby(['Foundation'])[['SalePrice']].agg(['mean','median','count'])
```
