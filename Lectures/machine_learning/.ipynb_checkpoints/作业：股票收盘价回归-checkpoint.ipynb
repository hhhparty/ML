## 准备数据

经过分析，股票数据中各字段的含义如下：

- date:日期
- ma5：5日平均收盘价
- ma10：10日平均收盘价
- ma20：20日平均收盘价
- v_ma5：5日平均成交量（单位是手，1手为100股）
- v_ma10：10日平均成交量
- v_ma20：20日平均成交量
- open：开盘价
- close：收盘价

下面的代码用于将数据文件读入内存，并以pandas.DataFrame对象存储。

下面使用线性回归进行预测。

在预测前必须将属性预处理，因为我选择的岭回归方法要求属性值必须经过归一化处理。

```python
from sklearn import preprocessing

X_scaled = preprocessing.scale(df.iloc[:,1:-1],axis = 0)
print(X_scaled)
y_scaled = preprocessing.scale(df.iloc[:,-1],axis = 0)
print(y_scaled)
```

归一化后，下面使用岭回归进行预测股价

```python
from sklearn.linear_model import Ridge
from sklearn import preprocessing
import numpy as np

reg = Ridge(alpha=1.0)
reg.fit(X_scaled[:-10], y_scaled[:-10]) 

print("学得岭回归模型的系数向量w为{}".format(reg.coef_))
print("学得岭回归模型的b为{}".format(reg.intercept_))
print("真实y：",y_scaled[-10:])
print("预测值y1={}".format(reg.predict(X_scaled[-10:])))

print("收盘价真值为：{}".format(df.iloc[-10:,-1].values.tolist()))
print("收盘价预测值为：{}".format(y_mean+reg.predict(X_scaled[-10:])*y_std))

from sklearn.metrics import mean_squared_error
print("均方误差为{}".format(mean_squared_error(y_scaled[-10:],reg.predict(X_scaled[-10:]))))

print("模型评分：",reg.score(X_scaled[-10:], y_scaled[-10:]))
```
